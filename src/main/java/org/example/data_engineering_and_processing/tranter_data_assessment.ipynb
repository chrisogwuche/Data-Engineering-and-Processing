```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('downloads/data_assessment.csv', delimiter=',');
```

```python
df.head()
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
plt.hist(df['value'], bins=10)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Value')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='type', y='value', data=df)
plt.title('Scatter plot of type vs value')
plt.show()
```
